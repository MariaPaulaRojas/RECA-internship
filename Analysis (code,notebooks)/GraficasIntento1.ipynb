```python
fig, ax = plt.subplots(1,3,figsize=(12, 3))
fig.tight_layout(pad=2)

parametros = ['MasaANegro', 'AcrecionMasaANegro', 'Masa']

for i_param, param  in enumerate(parametros):
    ax[i_param].scatter(First_table['DiscToTotal'], np.log10(First_table[param]),
                        color = 'gray', alpha=0.05)
    ax[i_param].set(xlabel='DiscToTotal', ylabel=param)
```
